```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Mall_Customers.csv")
X=data.iloc[:,[3,4]].values
```

```python
#using dendogram to find the optimal number of clusters
import scipy.cluster.hierarchy  as sch
dendogram=sch.dendrogram(sch.linkage(X,method='ward'))
#ward is used for minising the variants int the clusters
plt.title("DENDOGRAM")
plt.xlabel("customers")
plt.ylabel("euclidean distance")
plt.show()
```

```python
#seeing this we get 3 or 5 as optimal number of clusters so consider any!
```

```python
#training the hierarchial clustering mmodel on dataset
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5, affinity='euclidean',linkage='ward')
y_hc=hc.fit_predict(X)
```

```python
y_hc
```

```python
#visualising the clusters
plt.scatter(X[y_hc==0,0], X[y_hc==0,1], s=100,c='red', label='clus-1')
plt.scatter(X[y_hc==1,0], X[y_hc==1,1], s=100,c='blue', label='clus-2')
plt.scatter(X[y_hc==2,0], X[y_hc==2,1], s=100,c='pink', label='clus-3')
plt.scatter(X[y_hc==3,0], X[y_hc==3,1], s=100,c='green', label='clus-4')
plt.scatter(X[y_hc==4,0], X[y_hc==4,1], s=100,c='violet', label='clus-5')
plt.title("cluster of customers")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```
