```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
X=dataset.iloc[:,[3,4]].values
#there is no dependent variable in data set used for clustering
```

```python
#using elbow method to find the optimal number of clusters
# wcss-within clusters sum of squares

from sklearn.cluster import KMeans
#we will run for loop for 10 different no.of clusters and each time we run we will compute wcss and this will on y axis of the elbow method
wcss=[]
for i in range(1,11):
    kmeans= KMeans(n_clusters=i, init='k-means++',random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title("the elbow method clustering")
plt.xlabel("No.of Clusters")
plt.ylabel("WCSS")
plt.show()
        
```

```python
#from number 5 the curve decreases slowing and optimal no of clusteers is 5
```

```python
#training k-means model on the dataset
kmeans= KMeans(n_clusters=5, init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(X)
```

```python
print(len(y_kmeans))
```

```python
#visualizing the clusters
plt.scatter(X[y_kmeans==0,0], X[y_kmeans==0,1],s=100,c='red',label='clus-1')
plt.scatter(X[y_kmeans==1,0], X[y_kmeans==1,1],s=100,c='blue',label='clus-2')
plt.scatter(X[y_kmeans==2,0], X[y_kmeans==2,1],s=100,c='green',label='clus-3')
plt.scatter(X[y_kmeans==3,0], X[y_kmeans==3,1],s=100,c='yellow',label='clus-4')
plt.scatter(X[y_kmeans==4,0], X[y_kmeans==4,1],s=100,c='orange',label='clus-5')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300, c='cyan',label='centeroids')
plt.title("visualising k-means")
plt.xlabel("Annual Income")
plt.ylabel("spending score")
plt.legend()
plt.show()
```
